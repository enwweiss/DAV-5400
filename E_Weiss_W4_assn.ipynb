Question 3: Write a Python/Numpy code block that removes from array a any items that are also present in array b

```python
#Task 3
a = np.array([12, 5, 7, 15, 3, 1, 8])
b = np.array([14, 6, 3, 11, 19, 12, 5])

a[np.logical_not(np.in1d(a, b))]
```

Question 4:
Suppose you have ridden a bicycle from New York City to Key West, Florida. Your bicycle odometer shows the total miles you have travelled thus far, which you make a note of each day with paper and pencil. Your first two entries might be ‘55’ and ‘120’, indicating that you rode your bike 55 miles on day 1 and 65 miles on day 2. Your task is to create a NumPy array wherein you can record the cumulative miles you recorded each day during your trip. Then use your Python skills to show the total miles that you rode each day. Assume the trip took a total of 35 days and included at least 4 non-consecutive days where no cycling was possible due to either weather conditions or personal fatigue/soreness. For the remaining 31 days, you would have covered the entire distance between NYC and Key West. Complete your work within your Jupyter Notebook for this assignment and be sure to include a writeup explaining your approach to this problem, including the ways in which you decided to make use of NumPy.

My approach was to use the random function to generate a "day" of bike riding which can vary around the average amount of distance needed to be traveled over the 31 days of travel given. 

Data for how many miles was generated by google maps for a biking path, 130 miles were omitted from this schedule since they are taken on a ferry and are relativly quick. 

This method also omits days with 0 miles traveled since calculating this would go against the purpose of the program as it is designed to generate a random day of biking and on a break day there is no such trip.

Data is tracked in two independent methods: there is the "trip" which is the bike trip of that given day, and there is the "odimter" which is the running sum of the miles traveled.

A random "day" of bike riding was estimated as 1422/31 = 45 as the average, randomly generated days fall within a confidance interval around this average. The confidance interval was chosen through testing and is between 44 and 50 miles per day.



```python
#Task 4
import numpy as np

trip = []
days = 31
odmiter = []


for i in range(days):
    #if we already reach the destination we don't want
    # to keep creating new trips. to stop this I used an
    # if statement.
    if sum(trip)<1422:
        
        #Generate the bikeing trip for the day and 
        #store them in a variable
        miles_on_that_day = np.random.randint(44, 50)
        
        #Append the variable into a list "trip"
        trip.append(miles_on_that_day)
        
        #create the running sum for the miles traveled 
        # each day
        odmiter = [odmiter, sum(trip)]

print(sum(trip))
print(odmiter)
```
